### 3d annotate arrow

```python
from matplotlib.text import Annotation

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.proj3d import proj_transform
from mpl_toolkits.mplot3d.axes3d import Axes3D

class Annotation3D(Annotation):

    def __init__(self, text, xyz, *args, **kwargs):
        super().__init__(text, xy=(0, 0), *args, **kwargs)
        self._xyz = xyz

    def draw(self, renderer):
        x2, y2, z2 = proj_transform(*self._xyz, self.axes.M)
        self.xy = (x2, y2)
        super().draw(renderer)


def _annotate3D(ax, text, xyz, *args, **kwargs):
    '''Add anotation `text` to an `Axes3d` instance.'''

    annotation = Annotation3D(text, xyz, *args, **kwargs)
    ax.add_artist(annotation)

setattr(Axes3D, 'annotate3D', _annotate3D)

import numpy as np
from matplotlib.patches import FancyArrowPatch

class Arrow3D(FancyArrowPatch):

    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0, 0), (0, 0), *args, **kwargs)
        self._xyz = (x, y, z)
        self._dxdydz = (dx, dy, dz)

    def draw(self, renderer):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        super().draw(renderer)
        
    def do_3d_projection(self, renderer=None):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))

        return np.min(zs) 

def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add an 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)


setattr(Axes3D, 'arrow3D', _arrow3D)

```

```python
# digitalfilter.py

from collections import deque

import numpy as np


class LiveFilter:
    """Base class for live filters."""

    def process(self, x):
        # do not process NaNs
        if np.isnan(x):
            return x

        return _process(x)

    def __call__(self, x):
        return process(x)

    def _process(self, x):
        raise NotImplementedError("Derived class must implement _process")


class LiveLFilter(LiveFilter):
    """Live implementation of digital filter using difference equations.

    The following is almost equivalent to calling scipy.lfilter(b, a, xs):
    >>> lfilter = LiveLFilter(b, a)
    >>> [lfilter(x) for x in xs]
    """

    def __init__(self, b, a):
        """Initialize live filter based on difference equation.

        Args:
            b (array-like): numerator coefficients obtained from scipy
                filter design.
            a (array-like): denominator coefficients obtained from scipy
                filter design.
        """
        b = b
        a = a
        _xs = deque([0] * len(b), maxlen=len(b))
        _ys = deque([0] * (len(a) - 1), maxlen=len(a) - 1)

    def _process(self, x):
        """Filter incoming data with standard difference equations."""
        _xs.appendleft(x)
        y = np.dot(self.b, self._xs) - np.dot(self.a[1:], self._ys)
        y = y / self.a[0]
        self._ys.appendleft(y)

        return y


class LiveSosFilter(LiveFilter):
    """Live implementation of digital filter with second-order sections.

    The following is equivalent to calling scipy.sosfilt(sos, xs):
    >>> sosfilter = LiveSosFilter(sos)
    >>> [sosfilter(x) for x in xs]
    """

    def __init__(self, sos):
        """Initialize live second-order sections filter.

        Args:
            sos (array-like): second-order sections obtained from scipy
                filter design (with output="sos").
        """
        sos = sos

        n_sections = sos.shape[0]
        state = np.zeros((n_sections, 2))
        # print(x,x.shape)
        # state[state==0]=value1

    def _process(self, x):
        """Filter incoming data with cascaded second-order sections."""
        for s in range(n_sections):  # apply filter sections in sequence
            b0, b1, b2, a0, a1, a2 = sos[s, :]

            # compute difference equations of transposed direct form II
            y = b0 * x + state[s, 0]
            state[s, 0] = b1 * x - a1 * y + state[s, 1]
            state[s, 1] = b2 * x - a2 * y
            x = y  # set biquad output as input of next filter section.

        return y
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r"C:\Users\81809\Pictures\Camera Roll\WIN_20230522_03_36_33_Pro.jpg")

# Convert the image to the HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds of the green color in HSV
lower_green = np.array([40, 50, 50])
upper_green = np.array([80, 255, 255])

# Create a mask of the green pixels using the defined bounds
mask = cv2.inRange(hsv, lower_green, upper_green)

# Find contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Iterate over the contours and find the rectangular ones
rectangular_contours = []
for contour in contours:
    perimeter = cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, 0.04 * perimeter, True)
    if len(approx) == 4:
        rectangular_contours.append(approx)

# Draw rectangles around the identified rectangular contours
for contour in rectangular_contours:
    cv2.drawContours(image, [contour], -1, (0, 255, 0), 2)

# Display the image with the rectangles
cv2.imshow('Rectangular Boxes', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def compare_images( before_image, after_image):
    before_roi = cv2.imread(before_image)
    after_roi = cv2.imread(after_image)

    before_gray = cv2.cvtColor(before_roi, cv2.COLOR_BGR2GRAY)
    after_gray = cv2.cvtColor(after_roi, cv2.COLOR_BGR2GRAY)

    _, before_binary = cv2.threshold(before_gray, 127, 255, cv2.THRESH_BINARY)
    _, after_binary = cv2.threshold(after_gray, 127, 255, cv2.THRESH_BINARY)

    # Fill in ROI black and white image with white pixels less than 300
    before_binary[np.count_nonzero(before_binary == 255) < 300] = 0
    after_binary[np.count_nonzero(after_binary == 255) < 300] = 0

    before_black_pixels = np.count_nonzero(before_binary == 0)
    after_black_pixels = np.count_nonzero(after_binary == 0)
    pixel_difference = abs(after_black_pixels - before_black_pixels)

    inverted_before = cv2.bitwise_not(before_binary)
    inverted_after = cv2.bitwise_not(after_binary)

    return before_black_pixels, after_black_pixels, pixel_difference, inverted_before, inverted_after
```

```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def subtract_binary_images(image1, image2):
    # Ensure the images have the same shape
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same shape")

    # Perform binary subtraction
    result = cv2.subtract(image1, image2)

    return result

def threshold_roi(binary_image, threshold):
    binary_image[np.count_nonzero(binary_image == 255) < threshold] = 0
    return binary_image

def extract_angle_timestamp(image_filename):
    match = re.search(r'binary_inverted_(before|after)_(\d{8})_angle(\d+)\.png', image_filename)
    if match:
        angle = match.group(3)
        return angle
    else:
        return None

# Directory path for binary images
directory = r'C:\Users\81809\Desktop\Three_DoF_Robotarm-main\Three_DoF_Robotarm\utils\fig\binary_35_angle\binary_35'

# Get the list of binary image files in the directory
files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

# Sort the files in ascending order based on the angle value in the filename
files.sort(key=lambda x: int(extract_angle_timestamp(x)))

# Create a dictionary to store the pairs of before and after images
image_pairs = {}

# Load the before and after images and store them in the dictionary
for file in files:
    angle = extract_angle_timestamp(file)
    if angle is not None:
        if 'before' in file:
            image_pairs.setdefault(angle, {})
            image_pairs[angle]['before'] = cv2.imread(os.path.join(directory, file), cv2.IMREAD_GRAYSCALE)
        elif 'after' in file:
            image_pairs.setdefault(angle, {})
            image_pairs[angle]['after'] = cv2.imread(os.path.join(directory, file), cv2.IMREAD_GRAYSCALE)

# Create a slider widget
threshold_slider = widgets.IntSlider(
    value=300,
    min=0,
    max=1000,
    step=10,
    description='Threshold:',
    continuous_update=True,
    orientation='horizontal'
)

# Create a button widget
save_button = widgets.Button(description='Save')

# Create a label widget to display the pixel count
pixel_count_label = widgets.Label()

# Define the update function for the slider
def update_threshold(change):
    threshold = change['new']
    
    # Display the current image pair
    angle = next(iter(image_pairs.keys()))
    pair = image_pairs[angle]
    
    # Convert the images to binary
    _, before_binary = cv2.threshold(pair['before'], 127, 255, cv2.THRESH_BINARY)
    _, after_binary = cv2.threshold(pair['after'], 127, 255, cv2.THRESH_BINARY)
    
    # Threshold the ROIs
    before_thresholded = threshold_roi(before_binary.copy(), threshold)
    after_thresholded = threshold_roi(after_binary.copy(), threshold)
    
    # Update the pixel count label
    pixel_count_label.value = f'After {angle} (Pixel = {np.count_nonzero(after_thresholded == 255)})'
    
    # Display the images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(before_thresholded, cmap='gray')
    axes[0].set_title(f'Before {angle}')
    axes[0].axis('off')
    axes[1].imshow(after_thresholded, cmap='gray')
    axes[1].set_title(f'After {angle}')
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()

# Define the save button click event handler
def save_image(button):
    angle = next(iter(image_pairs.keys()))
    pair = image_pairs[angle]
    
    # Convert the images to binary
    _, after_binary = cv2.threshold(pair['after'], 127, 255, cv2.THRESH_BINARY)
    
    # Save the after thresholded image
    output_dir = os.path.join(directory, 'thresholded')
    os.makedirs(output_dir, exist_ok=True)
    output_filename = f"thresholded_{angle}.png"
    output_path = os.path.join(output_dir, output_filename)
    cv2.imwrite(output_path, after_binary)
    print(f"Image saved as {output_path}")
    
    # Remove the current image pair from the dictionary
    del image_pairs[angle]
    
    # Check if there are remaining image pairs
    if image_pairs:
        # Get the next angle
        next_angle = next(iter(image_pairs.keys()))
        
        # Update the slider value and trigger the update function
        threshold_slider.value = threshold_slider.min
        threshold_slider.observe(update_threshold, names='value')
        update_threshold({'new': threshold_slider.value})
    else:
        print("No more image pairs")

# Assign the save button click event handler
save_button.on_click(save_image)

# Display the slider widget
display(threshold_slider)

# Display the pixel count label
display(pixel_count_label)

# Display the save button widget
display(save_button)

# Initialize the first image pair
update_threshold({'new': threshold_slider.value})
```

```python
import cv2
import numpy as np

def fill_tapering_edges(image):
    # Find contours in the binary image
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a mask for filling the edges
    mask = np.zeros_like(image)

    # Iterate over each contour
    for contour in contours:
        # Fit a line to the contour using the least squares method
        rows, cols = image.shape[:2]
        [vx, vy, x, y] = cv2.fitLine(contour, cv2.DIST_L2, 0, 0.01, 0.01)

        # Calculate the start and end points of the line within the image bounds
        lefty = int((-x * vy / vx) + y)
        righty = int(((cols - x) * vy / vx) + y)

        # Draw the line on the mask
        cv2.line(mask, (cols - 1, righty), (0, lefty), 255, 1)

    # Fill the image using bitwise AND operation between the mask and the input image
    filled_image = cv2.bitwise_and(image, mask)

    return filled_image

# Load the binary image
binary_image = cv2.imread('binary_image.png', 0)

# Apply the tapering edges filling
filled_image = fill_tapering_edges(binary_image)

# Display the result
cv2.imshow('Filled Image', filled_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
